# Data Preprocessing Tools

## Importing the libraries

````
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
````

## Importing the dataset

````
#from typing import ValuesView
dataset = pd.read_csv('Data.csv')
# 2 new entities
#matrix of features - independent variables
#dependent variable-(usually last column)
X=dataset.iloc[:, :-1].values
y=dataset.iloc[:, -1].values



````

````
print(X)
````

````
print(y)
````

## Taking care of missing data

````
#can sometimes just delete 1% of data
#replace the data with average values
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(missing_values=np.nan, strategy='mean')
#choosing all values
imputer.fit(X[:,1:3])
#transforming the missing data
X[:,1:3]=imputer.transform(X[:,1:3])

````

````
print(X)
````

## Encoding categorical data

### Encoding the Independent Variable

````
#the countries are in names categorize them into 1 hot encoding 000,001,010 etc
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
#object column transform
ct=ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[0])] ,remainder='passthrough')
X=np.array(ct.fit_transform(X))
````

````
print(X)
````

### Encoding the Dependent Variable

````
#we will use label encoder to categorize no and yes (dependent varibales)
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y=le.fit_transform(y)
````

````
print(y)
````

## Splitting the dataset into the Training set and Test set

````
#creating 4 seperate sets which are pairs (training set and another one is matrix features)
# X train-matrix  features of training set
#X test matrix featues of test set
#Y train which is the dependent variable of training set
#Y test which is dependent variable of the test set
#training -> X test , Y test (input) -> fit methods ->prediction(inference)->predict the test
from sklearn.model_selection import train_test_split
X_train, X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)

````

````
print(X_train)
````

````
print(X_test)
````

````
print(y_train)
````
